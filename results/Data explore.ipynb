```python
%matplotlib inline
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import itertools
from scipy.stats import ks_2samp

sns.set(palette='colorblind')
```

```python
pm = pd.read_csv('JINT2020 - Summary.csv').sort_values(by=['Network', 'Width mult.', 'Input res.'])
pm['Inference time (i5 x 2)'] = 1000 / pm['Frame rate (i5 x 2)']
pm['Inference time (V100)'] = 1000 / pm['Frame rate (V100)']
pm['Inference time (Xeon x 48)'] = 1000 / pm['Frame rate (Xeon x 48)']
pm['Performance score (i5 x 2)'] = pm['mAP'] / pm['Frame rate (i5 x 2)']
pm['Performance score (V100)'] = pm['mAP'] / pm['Frame rate (V100)']
pm['Performance score (Xeon x 48)'] = pm['mAP'] / pm['Frame rate (Xeon x 48)']
pm['Normalized score (i5 x 2)'] = pm['Performance score (i5 x 2)'] / pm['Performance score (i5 x 2)'].max()
pm['Normalized score (V100)'] = pm['Performance score (V100)'] / pm['Performance score (V100)'].max()
pm['Normalized score (Xeon x 48)'] = pm['Performance score (Xeon x 48)'] / pm['Performance score (Xeon x 48)'].max()

v2_pm = pm[pm['Network'] == 'MobileNetV2']
v3_pm = pm[['MobileNetV3' in linha for linha in pm['Network']]]
yolo_pm = pm[['YOLO' in linha for linha in pm['Network']]]

pf = pd.read_csv('JINT2020 - FPS.csv')
pf['Inference time'] = 1000 / pf['Frame rate']

laptop_pf = pf[pf['Hardware Label'] == 'i5']
hpe_pf = pf[pf['Hardware Label'] != 'i5']
hpecpu_pf = pf[pf['Hardware Label'] == 'Xeon']
hpegpu_pf = pf[pf['Hardware Label'] == 'V100']
yolo_pf = pf[['YOLO' in linha for linha in pf['Network']]]

mobilenet_laptop_pf = laptop_pf[['MobileNet' in linha for linha in laptop_pf['Network']]]
mobilenet_cpu_pf = hpecpu_pf[['MobileNet' in linha for linha in hpecpu_pf['Network']]]
mobilenet_gpu_pf = hpegpu_pf[['MobileNet' in linha for linha in hpegpu_pf['Network']]]
```

```python
print(mobilenet_laptop_pf['Inference time'].mean(),
      mobilenet_laptop_pf['Inference time'].std(),
      mobilenet_gpu_pf['Inference time'].mean(),
      mobilenet_gpu_pf['Inference time'].std())
```

```python
print(yolo_pf.groupby(['Network', 'Hardware']).mean()['Inference time'].round(3))
print(yolo_pf.groupby(['Network', 'Hardware']).std()['Inference time'].round(3))
```

```python
print(hpe_pf.groupby(['Hardware', 'Network']).mean()['Inference time'].round(3))
# print(hpe_pf.groupby(['Hardware', 'Network']).std()['Inference time'].round(3))

# print(hpecpu_pf.groupby(['Network']).mean()['Inference time'].round(3) - hpegpu_pf.groupby(['Network']).mean()['Inference time'].round(3))

mob_mean_inf_cpu = mobilenet_cpu_pf['Inference time'].mean()
mob_mean_inf_gpu = mobilenet_gpu_pf['Inference time'].mean()
mob_std_inf_cpu = mobilenet_cpu_pf['Inference time'].std()
mob_std_inf_gpu = mobilenet_gpu_pf['Inference time'].std()

print(mob_mean_inf_cpu, mob_mean_inf_gpu, mob_mean_inf_cpu - mob_mean_inf_gpu)
print(mob_std_inf_cpu, mob_std_inf_gpu, mob_std_inf_cpu - mob_std_inf_gpu)
```

```python
print(len(mobilenet_cpu_pf['Inference time']), len(mobilenet_gpu_pf['Inference time']))
ks_2samp(mobilenet_cpu_pf['Inference time'], mobilenet_gpu_pf['Inference time'])
```

```python
for current_pf, current_hardware in (laptop_pf, 'i5-4210U@1.70GHz'), (hpecpu_pf, 'Xeon Gold 5118@2.3 GHz'), (hpegpu_pf, 'NVIDIA Tesla V100'):
    v2_pf = current_pf[current_pf['Network'] == 'MobileNetV2']
    v3_pf = current_pf[['MobileNetV3' in linha for linha in current_pf['Network']]]
    v3_pf['Network'].replace('MobileNet', 'MN',regex=True, inplace=True)
    v3_pf['Network'].replace('minimalistic', 'min.', regex=True, inplace=True)
    
    for comb in list(itertools.permutations(['Width mult.', 'Video res.', 'Input res.'], r=2)):
        g = sns.catplot(data=v2_pf, orient='v', height=5.5,aspect=1.3, kind='box', hue=comb[0], y="Inference time", x=comb[1], legend=False,)
        plt.tight_layout()
        plt.legend(loc='best',title=comb[0])
        g.fig.savefig('charts/v2_' + comb[0] + '_' + comb[1] + '_' + current_hardware + '.pdf')
    
    for comb in ['Width mult.', 'Video res.']:        
        g = sns.catplot(data=v3_pf, orient='v', height=5.5,aspect=1.1, kind='box', hue="Network", y="Inference time", x=comb, legend=False)
        plt.tight_layout()
        plt.legend(loc='best', title='Network')
        g.fig.savefig('charts/v3_' + comb + '_' + current_hardware + '.pdf')
```
